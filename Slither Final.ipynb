```python
import os
import re
import json # Import json module to work with JSON data
from slither.slither import Slither

folder_path = '/content'  # Replace with the actual folder path

for filename in os.listdir(folder_path):
    if filename.endswith('.sol'):
        file_path = os.path.join(folder_path, filename)
        output_filename = filename.replace('.sol', '.json')  # Generate output filename

        # Extract compiler version using a regular expression
        with open(file_path, 'r') as f:
            content = f.read()
        match = re.search(r'pragma solidity (\^?\d+\.\d+\.\d+);', content)
        if match:
            compiler_version = match.group(1)
            print(f"Found compiler version: {compiler_version} for file {filename}")

            # Install and use the extracted compiler version
            !pip install solc-select  # Install solc-select
            !solc-select install {compiler_version}
            !solc-select use {compiler_version}

        else:
            print(f"Compiler version not found in {filename}. Skipping.")
            continue  # Skip to the next file if compiler version is missing

        # Run Slither and save output to JSON file
        !slither {file_path} --json {output_filename}

        # Load the Slither results from the JSON file
        with open(output_filename, 'r') as f:
            data = json.load(f) # Load the JSON data into the variable 'data'

        # Iterate through detectors and print details
        if 'results' in data and 'detectors' in data['results']:
            for detector in data['results']['detectors']:
                print("\nDetector:")
                print("  Description:", detector['description'])
                print("  Impact:", detector['impact'])
                print("  Confidence:", detector['confidence'])

                if 'elements' in detector:
                    for element in detector['elements']:
                        print("    Element:")
                        print("      Type:", element['type'])
                        print("      Name:", element.get('name', ''))
                        print("      Source Mapping:", element['source_mapping'])

                        if 'type_specific_fields' in element:
                            for key, value in element['type_specific_fields'].items():
                                if isinstance(value, dict):
                                    print(f"      {key}:")
                                    for k, v in value.items():
                                        print(f"        {k}: {v}")
                                else:
                                    print(f"      {key}: {value}")
                        if 'additional_fields' in element:
                            for key, value in element['additional_fields'].items():
                                print(f"      {key}: {value}")
                else:
                    print("    No elements found for this detector.")
        else:
            print("Warning: 'detectors' key not found in the JSON result. Check Slither output format.")
```
